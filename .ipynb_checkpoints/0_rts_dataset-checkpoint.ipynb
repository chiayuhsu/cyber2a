## 1. Data

PyTorch provides two fundamental data primitives: `torch.utils.data.Dataset` and `torch.utils.data.DataLoader`, which streamline the process of handling data for machine learning tasks.

1. **Dataset**:
   - The `torch.utils.data.Dataset` is an abstract class that represents a dataset. Custom datasets can be created by subclassing `Dataset` and implementing two key methods:
     - `__len__`: Returns the size of the dataset.
     - `__getitem__`: Retrieves a data sample and its corresponding label based on an index.

2. **DataLoader**:
   - The `torch.utils.data.DataLoader` is a powerful utility that wraps an iterable around a `Dataset`. It facilitates easy access to data samples and supports key functionalities such as:
     - **Batching**: Automatically divides the data into batches of a specified size.
     - **Shuffling**: Randomly shuffles the data at every epoch, which helps to ensure that the model does not learn any unintended order.
     - **Parallel Loading**: Utilizes multiple workers to load data in parallel, improving the efficiency and speed of data loading.
   - This abstraction simplifies the process of iterating over data during training and evaluation.

3. **Transforms**:
   - PyTorchs `torchvision.transforms` module provides a variety of common image transformations that can be applied to datasets for preprocessing and data augmentation.
   - Transforms can convert raw data to a format suitable for training, such as:
     - **Normalization**: Adjusting the scale of input data to have a mean of 0 and a standard deviation of 1, which can improve model convergence.
     - **Resizing and Cropping**: Altering the size of images to a uniform dimension.
     - **Data Augmentation**: Applying random transformations like rotation, flipping, and color jitter to increase dataset diversity and help prevent overfitting.
   - Transforms can be composed using `transforms.Compose` to apply multiple preprocessing steps to the dataset in a sequential manner.

Together, `Dataset`, `DataLoader`, and `Transforms` form the backbone of data handling in PyTorch, enabling efficient and flexible data processing pipelines for deep learning applications.

### Dataset Introduction

In this tutorial, we will utilize the RTS dataset located in the `cyber2a` folder. The RTS dataset is a multi-modality dataset originally designed for segmentation tasks. However, for the purposes of this tutorial, we will convert it into a classification task to determine the number of RTS present in each image. The number of RTS ranges from 1 to 10, allowing us to use the count directly as the label.

The dataset includes three types of image modalities: RGB, NDVI, and NIR, providing diverse data sources to enhance classification performance. The folder structure and required files for this tutorial are as follows:

```
cyber2a
--- rts
    --- rgb  # RGB image folder
        --- train_nitze_000.jpg
        --- train_nitze_001.jpg
        --- ...
    --- ndvi # NDVI image folder
    --- nir  # NIR image folder
--- data_split.json
--- rts_cls.json
```

- The `data_split.json` file is a dictionary containing two keys: `train` and `valtest`.
  - The `train` key holds a list of image names designated for training.
  - The `valtest` key comprises a list of image names for validation and testing.

- The `rts_cls.json` file is a dictionary where each key corresponds to an image name, and the associated value is the number of RTS in that image.

In the following sections, we will demonstrate how to load an rgb image and its corresponding label from the dataset.

````python
import os
import json 

from PIL import Image
import matplotlib.pyplot as plt

# Define the directory where images are stored
img_dir = "cyber2a/rts/rgb/"

# Load the data split file to get lists of training and validation/test images
with open("cyber2a/rts/data_split.json", 'r') as f:
    data_split = json.load(f)

# Retrieve the list of training images
img_list = data_split["train"]

# Load the image labels, where each image name maps to the number of RTS in the image
with open("cyber2a/rts/rts_cls.json", 'r') as f:
    img_labels = json.load(f)

# Select the first image file name from the training list and get its corresponding label
img_name = img_list[0]
img_label = img_labels[img_name]

# Print the image file name and its corresponding number of RTS
print(f"Image Name: {img_name}, Number of RTS: {img_label}")

# Construct the full path to the image file
img_path = os.path.join(img_dir, img_name)

# Open the image and convert it to RGB format
image = Image.open(img_path).convert("RGB")

# Convert the label to 0-indexed format for classification tasks
label = int(img_label) - 1

# Display the image using matplotlib
fig, ax = plt.subplots()
ax.imshow(image)
ax.axis('off')  # Hide the axis
ax.set_title(f"Label: {label}")  # Set the title to the 0-indexed label

# Show the plot
plt.show()
````
